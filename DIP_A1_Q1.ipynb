```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img



def histogram(img):
    
    img_array = np.array(img)
    plt.gray()
    plt.axis('off')
    plt.imshow(img_array)
    plt.show()
    
    dim1, dim2 = np.shape(img_array)
    print("Actual average intensity", np.sum(img_array)/(dim1*dim2))
    hist = np.zeros(256)
    
    #Computing the Histogram of Intensities
    for i in range(0, dim1):
        for j in range(0, dim2):
            pixel_value = img_array[i][j]
            hist[pixel_value] = hist[pixel_value] + 1
    
    
    sum = 0
    for k in range(0, 256):
        sum = sum + k*hist[k]
    print('\nThe histogram of intensities of the image is\n ', hist)
    
    #Plotting Histogram of Intensities
    x = np.linspace(0,256,256)
    plt.xlabel("Intensity levels")
    plt.ylabel("Frequency of occurence")
    plt.plot(x,hist)
    plt.show
    return hist, dim1, dim2, sum

def main():
    
    #Load the image
    img = Image.open('coins.png')
    
    hist, dim1, dim2, sum = histogram(img)
    print("\nAverage intensity calculated from Histogram", sum/(dim1*dim2))
    
    return 1

if __name__ == "__main__":
    main()


   
```
